Reading the data:

```
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
with open('marital_sex_district_query.csv', 'r') as file:
  reader = csv.reader(file)
  date = []
  birth = []
  for row in reader:
    if row[0] != 'REF_YEAR':
      date.append(int(row[0]))
    if row[1] != 'LIVEBIRTH':
      birth.append(int(row[1]))
  print(date, birth)
```

Printing the data:

```
s = pd.Series(birth, index=date)
s.plot()
plt.title('Entwicklung der Geburtenrate')
plt.show()
```

Reading the data:

```
with open('marital_born_query.csv', 'r') as file:
  reader = csv.reader(file)
  mar_sum = 0
  mar_n_sum = 0
  for row in reader:
    if row[0] != 'MARITAL':
      mar_sum += int(row[0])
    if row[1] != 'NON_MARITAL':
      mar_n_sum += int(row[1])
  print(mar_sum, mar_n_sum)
  
```

Printing the data:

```
plt.bar(['marital', 'non-marital'], [mar_sum, mar_n_sum])
plt.title('Mütter, die verheiratet / unverheiratet bei der Geburt waren')
plt.show()
```

Reading the data:

```
with open('gender_after_district.csv', 'r') as file:
  reader = csv.reader(file)
  gender_male_one = 0
  gender_female_one = 0
  gender_male_two = 0
  gender_female_two = 0
  gender_male_three = 0
  gender_female_three = 0
  for row in reader:
    if row[0] != 'gender_male_one':
      gender_male_one += int(row[0])
    if row[1] != 'gender_female_one':
      gender_female_one += int(row[1])
    if row[2] != 'gender_male_two':
      gender_male_two += int(row[2])
    if row[3] != 'gender_female_two':
      gender_female_two += int(row[3])
    if row[4] != 'gender_male_three':
      gender_male_three += int(row[4])
    if row[5] != 'gender_female_three':
      gender_female_three += int(row[5])
  print(f'gm/1: {gender_male_one}, gf/1: {gender_female_one}, gm/2: {gender_male_two}, gf/2: {gender_female_two}, gm/3: {gender_male_three}, gf/3: {gender_female_three}')
```

Printing the data:

```
fig = plt.subplots(figsize=(10, 7))
male = plt.bar(np.arange(3), (gender_male_one, gender_male_two, gender_male_three), 1)
female = plt.bar(np.arange(3), (gender_female_one, gender_female_two, gender_female_three), 1)
plt.xticks(np.arange(3), ('District 1', 'District 2', 'District 3'))
#plt.yticks(np.arange(180000))
plt.legend((male[0], female[0]), ('männlich', 'weiblich'))
plt.show()
```
